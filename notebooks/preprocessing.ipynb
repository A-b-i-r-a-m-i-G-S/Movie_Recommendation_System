```python
import numpy as np
import pandas as pd
```

```python
movies_df = pd.read_csv('../datasets/imdb_movies.csv')
movies_df.head()
```

```python
movies_df.info()
```

```python
movies_df = movies_df.dropna(how='any')
```

```python
def make_genresList(x):
    genres = []
    for item in x.split(","):
        item = item.strip()
        if item=='Science Fiction':
            item = 'Sci-Fi'
        genres.append(item)
    
    if genres == []:
        return np.NaN
    return " ". join(genres)        

movies_df['genres'] = movies_df['genre'].map(lambda x: make_genresList(x))
```

```python
def get_actor_names(x):
    actors = []
    for item in x.split(",")[:3]:
        item = item.strip()
        actors.append(item)
    
    if actors == []:
        return np.NaN
    return " ". join(actors)        

movies_df['actors'] = movies_df['crew'].map(lambda x: get_actor_names(x))
```

```python
movies_df.head()
```

```python
movies_df.drop_duplicates(subset ="names", keep = 'last', inplace = True)
```

```python
movies_df['content'] = movies_df['names'] + " " + movies_df['genres'] + " " + movies_df['overview'] + " " + movies_df['actors']
```

```python
movies_df['title'] = movies_df['names'].str.lower()
```

```python
movies_dataset = movies_df.loc[:, ['title','score','content']]
movies_dataset.head()
```

```python
movies_dataset.to_csv('../datasets/movies.csv',index=False)
```
